```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

data=pd.read_csv('/content/Sales.csv')
```

```python
x= data['GrLivArea']
y= data['SalePrice']

x=(x-x.mean())/x.std()
x=np.c_[np.ones(x.shape[0]),x]
```

```python
alpha=0.01
iterations=2000
m=y.size
np.random.seed(123)
theta=np.random.rand(2)

def gradient_descent(x,y,theta,iterations,alpha):
  past_costs=[]
  past_thetas=[theta]
  for i in range(iterations):
    prediction=np.dot(x,theta)
    error=prediction-y
    cost=(1/(2*m))*np.dot(error.T,error)
    past_costs.append(cost)
    theta=theta-(1/m)*alpha*(np.dot(x.T,error))
    past_thetas.append(theta)
  return past_thetas,past_costs

past_thetas,past_costs=gradient_descent(x,y,theta,iterations,alpha)
theta=past_thetas[-1]
print("Gradint Desent:{:.2f},{:.2f}".format(theta[0],theta[1]))
```

```python

plt.title('Cost Function J')
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.plot(past_costs)
plt.show()
```

```python
from __future__ import annotations
fig = plt.figure()
ax= plt.axes()
plt.title('Sale Price vs Living Area')
plt.xlabel('Living Area in square feet (normalised)')
plt.ylabel('Sale Price ($)')
plt.scatter(x[:,1],y,color='red')
line, =ax.plot([],[],lw=2)
annotations=ax.text(-1 , 700000, ' ')
annotations.set_animated(True)
plt.close()

def init():
  line.set_data([],[])
  annotations.set_text(' ')
  return line,annotations

def animate(i):
  x=np.linspace(-5,20,1000)
  y=past_thetas[i][1]*x+past_thetas[i][0]
  line.set_data(x,y)
  annotations.set_text('Cost = %.2f e10' % (past_costs[i]/10000000000))
  return line,annotations

anim=animation.FuncAnimation(fig,animate,init_func=init,frames=2000,interval=0,blit=True)
anim.save('animation.gif',writer='imagemagick',fps = 30)
```
